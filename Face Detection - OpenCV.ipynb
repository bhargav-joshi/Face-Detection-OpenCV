```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = plt.imread("./group-photo.jpg")
```

```python
plt.imshow(img)
```

```python
detector = cv2.CascadeClassifier("haarcascade_frontalface_alt (2).xml")
```

```python
faces = detector.detectMultiScale(img)
```

```python
faces
```

```python
faces.shape
```

```python
faces[0]
```

```python
x, y, w,h = faces[0]
```

```python
x
```

```python
y
```

```python
w
```

```python
h
```

```python
img = cv2.rectangle(img, (x, y) , (x+w, y+h), (255, 255, 255), 2 )
```

```python
plt.imshow(img)
```

## Plotting Rectangles For All Faces

```python
for face in faces:
    x,y,w,h = face
    img = cv2.rectangle(img, (x, y) , (x+w, y+h), (255, 255, 255), 3 )
```

```python
plt.imshow(img)
```
